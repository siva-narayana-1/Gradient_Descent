```python
import numpy as np
import pandas as pd
```

```python
x = np.array([1,2,3,4,5], dtype=float)
y = np.array([55,56,57,59,61], dtype=float)
```

```python
n = len(y)
w = 0.0
b = 0.0
lr = 0.01

for epoch in range(1000):
    y_pred = w*x + b

    dw = (-2/n) * np.sum((y - y_pred) * x)
    db = (-2/n) * np.sum(y - y_pred)

    if abs(dw) < 0:
        print("Negative touched...")
        break

    w -= lr * dw
    b -= lr * db

print("w value is:", w)
print("b value is:", b)
```
